```python
from sklearn.datasets import load_iris
```

```python
# Loading the dataset
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y
```

```python
# first five rows
df.head()
```

```python
df['target'].unique()
```

```python
# visuliazing the dendogram
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

linked = linkage(X,method='ward')
dendrogram(linked,orientation='top',distance_sort='descending',show_leaf_counts=True)
plt.title("hirearchy clustering")
plt.xlabel("samples")
plt.ylabel("distance")
plt.show()
```

```python
# Training the model
k = 2
model = AgglomerativeClustering(n_clusters=k,metric='euclidean',linkage='ward')
df['cluster'] = model.fit_predict(X)
```

```python
df.head()
```

```python
# visualizing the clusters
plt.figure(figsize=(8, 6))

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=df['cluster'], palette='Set2', s=100)

plt.title('Agglomerative Clustering (First Two Features)')

plt.xlabel(iris.feature_names[0])

plt.ylabel(iris.feature_names[1])

plt.grid(True)

plt.legend()

plt.show()
```

```python
# visualizing the clusters
plt.figure(figsize=(8, 6))

sns.scatterplot(x=X[:, 2], y=X[:, 3], hue=df['cluster'], palette='Set2', s=100)

plt.title('Agglomerative Clustering (next  Two Features)')

plt.xlabel(iris.feature_names[2])

plt.ylabel(iris.feature_names[3])

plt.grid(True)

plt.legend()

plt.show()
```
