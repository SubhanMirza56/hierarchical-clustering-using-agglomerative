```python
from sklearn.datasets import load_iris
```

```python
import pandas as pd

iris = load_iris()
X = iris.data
y = iris.target
df = pd.DataFrame(X, columns=iris.feature_names)
df.head()
```

Applying Elbow method to find the value of k

```python
import matplotlib.pyplot as plt
from sklearn.cluster import BisectingKMeans
sse = []
K_range = range(1, 11)
for k in K_range:
    model = BisectingKMeans(n_clusters=k, random_state=0)
    model.fit(X)
    sse.append(model.inertia_)  # Inertia = total SSE

# Plot elbow
plt.plot(K_range, sse, marker='o')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (SSE)")
plt.title("Elbow Method for Choosing k")
plt.grid()
plt.show()
```

Since the elbow is making at point 2 so we select the value of k = 2

we are using BisectingKmeans algorithum

```python
from sklearn.cluster import BisectingKMeans

df['cluster'] = BisectingKMeans(n_clusters=2, random_state=0).fit_predict(X)
```

```python
df.head()
```

```python
# visualizing our clusters on first two features
import seaborn as sns
plt.figure(figsize=(8, 6))

sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=df['cluster'], palette='Set2', s=100)

plt.title('Divisive Clustering (First Two Features)')

plt.xlabel(iris.feature_names[0])

plt.ylabel(iris.feature_names[1])

plt.grid(True)

plt.legend()

plt.show()
```

```python
# visualizing our clusters on last two features

import seaborn as sns
plt.figure(figsize=(8, 6))

sns.scatterplot(x=X[:, 2], y=X[:, 3], hue=df['cluster'], palette='Set2', s=100)

plt.title('Divisive Clustering (First Two Features)')

plt.xlabel(iris.feature_names[2])

plt.ylabel(iris.feature_names[3])

plt.grid(True)

plt.legend()

plt.show()
```
